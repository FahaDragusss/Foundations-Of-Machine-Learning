```python
import pandas as pd
import numpy as np
import matplotlib as plt
%matplotlib inline
```

```python
house_df = pd.read_csv("house_price_dataset.csv")
house_df
```

```python
house_df.plot.scatter(x = 'price', y= 'bathrooms')
house_df.plot.scatter(x = 'price', y= 'bedrooms')
```

```python
house_df.describe()
```

```python
df_filtered = house_df.drop(['waterfront','view','sqft_basement','sqft_above','floors','condition',
                             'street','yr_renovated','city','statezip','country','date','yr_built',
                             'sqft_lot','sqft_living'], axis= 1)
df_filtered
```

```python
house_filtered = df_filtered[
    (df_filtered['price'] >= 1) &
    (df_filtered['bedrooms'] >= 1) &
    (df_filtered['bathrooms'] >= 1)
]
house_filtered.insert(0,'X0', 1)
# house_filtered['age'] = 2014 - house_df['yr_built']
house_filtered
```

```python
house_filtered.describe()
```

```python
house_filtered.to_csv('house_price_dataset_handled.csv', index = False)
```
